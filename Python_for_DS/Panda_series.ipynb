```python
#task1 create unique student id for 40 students
import numpy as np
import pandas as pd
studid=pd.Series(np.arange(1001,1041))
studid
```

```python
#task2 create series for maths & physics with student id as index
maths_marks=[35, 86, 63, 90, 50, 63, 15, 72, 15, 54, 72, 83, 58, 70, 92, 40, 59, 63, 53, 59, 55, 85, 18, 83, 67, 85, 46, 59, 77, 71, 77, 56, 79, 61, 82, 76, 60, 52, 88, 78]
physics_marks=[85, 94, 63, 59, 76, 64, 87, 55, 51, 71, 54, 94, 33, 72, 75, 67, 85, 15, 50, 50, 19, 55, 59, 65, 76, 13, 57, 70, 51, 94, 55, 50, 71, 91, 75, 51, 90, 70, 87, 56]
math_series=pd.Series(maths_marks, index=studid)
phy_series=pd.Series(physics_marks, index=studid)
print(math_series)
print(phy_series)
```

```python
#task3 create series of total marks from maths & physics with student id as index
total_series=pd.Series(math_series+phy_series, index=studid)
print('Series with student id & total score from math & physics:')
print(total_series)
```

```python
#task4 students with total marks above 120
print('Students qualified for round1:')
round1=total_series[total_series>=120]
print(round1.index)
```

```python
#task5 create data frame to store round 1 qualifiers
dict1={'Round1 qualifiers':round1.index}
df_round1=pd.DataFrame(dict1)
df_round1
```

```python
#extra: to store all details in a data frame
dictn={'Maths':math_series.values,'Physics':phy_series.values,'Total':total_series.values}
student_info=pd.DataFrame(dictn,index=studid)
student_info
```

```python
#task6 create data frame to store round 2 qualifiers
filter_data=student_info[(student_info.Maths>=50)&(student_info.Physics>=50)&(student_info.Total>=120)]
filter_data.index
df_round2=pd.DataFrame(filter_data.index,columns=['Round2 qualifiers'])
df_round2
```

```python
#task7 count of students who qualify round 2
print('The number of students who qualify for round 2:',df_round2.size)
```
