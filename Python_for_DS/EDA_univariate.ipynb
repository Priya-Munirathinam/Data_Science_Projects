```python
#task 2: function that identifies type of column
def data_type(col):
    if df[col].dtypes!='object':
        print(f'The {col} column contains values of numerical type')
    else:
        print(f'The {col} column contains values of categorical type')

data_type('JobInvolvement')
```

```python
c_list=[]
n_list=[]
for col in df.columns:
    if df[col].dtypes!='object':
        n_list.append(col)
    else:
        c_list.append(col)

print(tabulate({'categorical':c_list,'Numerical':n_list},headers=['Categorical','Numerical']))  
```

```python
#Task 3: information of columns
def categ_info(col):
    print(f'unique values in {col} are: {df[col].nunique()}') #unique will return distinct values in column
    print(f'mode function for {col} are: {df[col].mode()[0]}') # return mode function & its yop most elt
    print(f'no of missing values in {col} are: {df[col].isnull().sum()}')
```

```python
#Task4: descreptive analysis on categorical columns
def Cat_count_plot(col):
    fig,ax=plt.subplots(1,2,figsize=(15,7))
    ax[0].set_title(f'Count of {col}')
    prcn=df[col].value_counts()
    lab=df[col].value_counts().index

    sns.countplot(x=df[col],ax=ax[0])
    plt.pie(prcn,labels=lab,autopct='%.2f%%')
    plt.show()
```

```python
# Task 5: function to print information of numerical column
def num_info(col):
    print(f'mean of {col} is: {df[col].mean()}')
    print(f'median of {col} is: {df[col].median()}')
    print(f'Standard deviation of {col} is: {df[col].std()}')
    print(f'no of missing values in {col} are: {df[col].isnull().sum()}')
    print(f'Skew of {col} is: {skew(df[col])}')
```

```python
num_info('Age')
```

```python
#Task 6: Descriptive analysis on Numerical columns
def Num_count_plot(clm):
    fig,ax=plt.subplots(1,2, figsize=(12,5),gridspec_kw={'width_ratios':(0.35,0.75)})
    ax[0].set_title(f'Boxplot for {clm}')
    ax[1].set_title(f'Histogram for {clm}')
    plt.hist(df[clm],rwidth=0.98,color='grey')
    sns.boxplot(df[clm],ax=ax[0],color='orange')
    plt.show()
```
