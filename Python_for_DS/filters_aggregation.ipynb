Task 1: Treating the missing values in given data set

```python
#2.identify missing value columns
df.isnull().sum()
```

```python
# mode for categorical 
df['gender'].value_counts()
```

```python
#impute for gender column using mode
df['gender']=df['gender'].fillna('Male')
```

```python
#check for missing vales
df['gender'].isnull().sum()
```

```python
# For test1 & test2 columns using mean & median respectively
df['test1']=df['test1'].fillna(df['test1'].mean())
df['test2']=df['test2'].fillna(df['test2'].median())
```

```python
df
```

```python
df.isnull().sum() # now all missing values are imputed
```

```python
#task2: display the duplicate rows if any
df[df.duplicated()] #no rows are duplicates
```

```python
#task3: replace the values in column 'tuitionfee_waiver' with no or yes 
df.loc[df['tuitionfee_waiver']=='Does not qualify','tuitionfee_waiver']='No'
df.loc[df['tuitionfee_waiver']=='Qualifies for reduced/free ','tuitionfee_waiver']='Yes' 
```

```python
#task4: groupby to display average score of test1, schoolwise
avg_test1 = df.groupby(['school'])[['test1']].mean()
avg_test1
```
