```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('DS1_C4_S5_Employee_Data_Practice.xlsx')
data
```

```python
#task1: departtment wise salary in bar chart
#a. group by department
df1= data.groupby(['Department'])[['Annual Salary ($)']].sum().sort_values('Annual Salary ($)')
df1
```

```python
#b. barchart
plt.barh(df1.index,df1['Annual Salary ($)']) 
plt.title('Department wise total annual salary')
plt.xlabel('Total salary')
plt.ylabel('Departments')
plt.show()
```

```python
#task2: data frame to display gender wise total number of employees in each department 
df2 = data.groupby(['Department','Gender'])[['Gender']].count()
df2
```

```python
#task3: The salary of employees with no work experience in IT department is filtered & average is calculated
df3=data[(data['Department']=='IT ')&(data['Work_Experience']==0)] 
print('The salary, company might offer new employee with no work experience in IT department is:')
df3['Annual Salary ($)'].mean()
```

```python
#task4: pie chart for department wise annual salary
plt.pie(df1['Annual Salary ($)'],labels=df1.index,autopct='%.2f%%') 
plt.title('Department wise annual salary composition')
plt.show()
```

```python
#task5: histogram to display the distribution of employee count by age group
bin=[20,25,30,35,40,45,50]
plt.hist(data['Age'],bin,rwidth=0.98)
plt.title('Distribution of age of employees')
plt.xlabel('Age group')
plt.ylabel('No of employees')
plt.show()
```

```python
#task6: salary trend based on employee experience
plt.scatter(data['Work_Experience'],data['Annual Salary ($)'],color='y',s=30)
plt.title('salary trend based on experience')
plt.xlabel('Experience')
plt.ylabel('Annual Salary')
plt.show()
```

```python
#task7: Box plot for department wise salary
data.boxplot(by='Department',column=['Annual Salary ($)'],figsize=(14,5))
#IT department has outlier above 160000
#HR deaprtment has outlier 140000
```
