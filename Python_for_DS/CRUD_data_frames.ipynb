```python
import pandas as pd

import warnings
warnings.filterwarnings='ignore'
```

```python
#task1 data in ascending order of sales made
Salesdata=pd.read_csv('DS1_C4_S3_SalesData_Practice.csv')
Salesdata
print('The data of sales in ascending order:')
Salesdata.sort_values('Sales')
```

```python
#task2 data where sales made is greater than or equal to 45000
Salesdata[Salesdata.Sales>=45000]
```

```python
#task3: data det for california, texas, montana & display each in ascending order of sales
#california
Cal=Salesdata[Salesdata.State=='California'].sort_values('Sales')
print('Sales in ascending order for California:')
Cal
```

```python
#Montana
Mont=Salesdata[Salesdata.State=='Montana'].sort_values('Sales')
print('Sales in ascending order for Montana:')
Mont
```

```python
#Texas
Tex=Salesdata[Salesdata.State=='Texas'].sort_values('Sales')
print('Sales in ascending order for Texas:')
Tex
```

```python
#task4 sum of sales for 3 states: california, montana, texas
print('The total sales made in California:',Cal.Sales.sum())
print('The total sales made in Montana:',Mont.Sales.sum())
print('The total sales made in Texas:',Tex.Sales.sum())
```

```python
#task5: the month with highest & lowest sales in california
maxm=Cal.iloc[-1]['Month']
minm=Cal.iloc[0]['Month']
print('For California:')
print('The Maximum sales was made in the month:',maxm)
print('The Minimum sales was made in the month:',minm)
```

```python
#pre requisite
N_Mex=Salesdata[Salesdata.State=='New Mexico']
Alask=Salesdata[Salesdata.State=='Alaska']
N_Mex.Sales.sum()
Alask.Sales.sum()
```

```python
#task 6: Create a data frame on each state wise sales
dict={'State':['Alaska','Texas','California','Montana','New Mexico'],'Sales in thousands':[Alask.Sales.sum(),Tex.Sales.sum(),Cal.Sales.sum(),Mont.Sales.sum(),N_Mex.Sales.sum()]}
state_sale=pd.DataFrame(dict)
state_sale
```

```python
#task7: sort the task6 dataframe and save into a csv file
sort_sale=state_sale.sort_values(['Sales in thousands'])
sort_sale.to_csv('total_sales.csv',index=False)
```
