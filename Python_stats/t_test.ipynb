```python
import pandas as pd
import numpy as np
import statistics as st
import math
from scipy.stats import t
import matplotlib.pyplot as plt
```

# Task 1

```python
#3) set alpha:
alpha = 0.1
#4) decision rule:
# if p-value < alpha & -z_crit > z_stat > z_crit --> reject null hypothesis
```

```python
#5) analyse data
p_mean = 71
score = [61,93,75,67,84,95]
s_mean = st.mean(score)
s_std = st.stdev(score)
n = 6
df = 5 # 6 - 1
print('Assumed population mean',p_mean)
print('Mean of smaple',s_mean)
print('Std dev:',s_std)
print('sample size',n)
print('degree of freedom',df)
```

```python
#6) statistical parameter 
# task 1c
t_statistic = (s_mean - p_mean) / (s_std/math.sqrt(n)) 
print("The t-statistic is ", t_statistic)

p_value = t.sf(abs(t_statistic),df)
print("The p_value is ", p_value)

t_critical = t.ppf(1-alpha, df)
print("The t-critical value is "+str(t_critical)) # task 1b
```

```python
# task 1d
x = np.arange(-4,4,0.1)
tvalues5 = t.pdf(x,5)

plt.plot(tvalues5)
plt.legend(['df = 5'])
plt.show()
```

# Task 2

```python
#3) set alpha:
alpha = 0.1
#4) decision rule:
# if p-value < alpha & -z_crit > z_stat > z_crit --> reject null hypothesis
```

```python
#5) analyse data
batch1 = [9,9,9,36,45,48,51,57,69,72,96]
x1 = st.mean(batch1)
v1 = st.variance(batch1)
n1 = 11
batch2 = [60,39,39,60,79,96,69,60,75,45,90]
x2 = st.mean(batch2)
v2 = st.variance(batch2)
n2 = 11
df = 20 # 11 + 11 -2
print('Batch 1:')
print('Mean',x1)
print('Variance:',v1)
print('sample size',n1)
print('--------')
print('Batch 2:')
print('Mean',x2)
print('Variance:',v2)
print('sample size',n2)
print('-------')
print('degree of freedom',df)
```

```python
#6) statistical parameter
# task 2c
noise = math.sqrt(v1*(n1-1) + v2*(n2-1))*math.sqrt(1/n1 + 1/n2)
signal = abs(x1-x2)
t_statistic=signal/noise
print('t-statistic =',t_statistic)

p_value=t.sf(abs(t_statistic),df)*2
print("The p_value is ", p_value)

t_critical = t.ppf(1-alpha/2, df) # task 2b
print('t-critical =',t_critical)
```
