```python
data=pd.read_excel('PRO_Car_Data.xlsx')
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
#filling 'Make' column with 'Not available'
data['Make']=data['Make'].fillna('Not Available')
#filling 'Body_Type' column with 'Not available'
data['Body_Type']=data['Body_Type'].fillna('Not Available')
#filling 'Drive_Modes' column with 'Not available'
data['Drive_Modes']=data['Drive_Modes'].fillna('Not Available')
#filling 'Displacement' column with mean
data['Displacement']=data['Displacement'].fillna(data['Displacement'].mean())
#filling 'Fuel_Tank_Capacity_litre' column with mean
data['Fuel_Tank_Capacity_litre']=data['Fuel_Tank_Capacity_litre'].fillna(data['Fuel_Tank_Capacity_litre'].mean())
#filling 'Valves_Per_Cylinder' column with mean
data['Valves_Per_Cylinder']=data['Valves_Per_Cylinder'].fillna(data['Valves_Per_Cylinder'].mean())
#filling 'Fuel_Tank_Capacity_litre' column with mean
data['Fuel_Tank_Capacity_litre']=data['Fuel_Tank_Capacity_litre'].fillna(data['Fuel_Tank_Capacity_litre'].mean())
#filling 'Torque' column with mean
data['Torque']=data['Torque'].fillna(data['Torque'].mean())
#filling 'Cylinders' column with mean
data['Cylinders']=data['Cylinders'].fillna(data['Cylinders'].mean())
data.isnull().sum()
```

```python
#filling 'ABS_(Anti-lock_Braking_System)' column with 'Not available'
data['ABS_(Anti-lock_Braking_System)']=data['ABS_(Anti-lock_Braking_System)'].fillna('Not Available')
#filling 'Airbags' column with 'Not available'
data['Airbags']=data['Airbags'].fillna('Not Available')
#filling 'Hill_Assist' column with 'Not available'
data['Hill_Assist']=data['Hill_Assist'].fillna('Not Available')
#filling 'Central_Locking' column with 'Not available'
data['Central_Locking']=data['Central_Locking'].fillna('Not Available')
#filling 'Child_Safety_Locks' column with 'Not available'
data['Child_Safety_Locks']=data['Child_Safety_Locks'].fillna('Not Available')
#filling 'High_Speed_Alert_System' column with 'Not available'
data['High_Speed_Alert_System']=data['High_Speed_Alert_System'].fillna('Not Available')
#filling 'Fasten_Seat_Belt_Warning' column with 'Not available'
data['Fasten_Seat_Belt_Warning']=data['Fasten_Seat_Belt_Warning'].fillna('Not Available')
#filling 'Door_Ajar_Warning' column with 'Not available'
data['Door_Ajar_Warning']=data['Door_Ajar_Warning'].fillna('Not Available')
#filling 'Torque' column with mean
data['Torque']=data['Torque'].fillna(data['Torque'].mean())
#filling 'Seating_Capacity' column with mean
data['Seating_Capacity']=data['Seating_Capacity'].fillna(data['Seating_Capacity'].mean())
#filling 'Height_mm' column with mean
data['Height_mm']=data['Height_mm'].fillna(data['Height_mm'].mean())
#filling 'Width_mm' column with mean
data['Width_mm']=data['Width_mm'].fillna(data['Width_mm'].mean())
data.isnull().sum()
```

```python
#replace 'ABS_(Anti-lock_Braking_System)' column with 'yes' as 1 and'Not available' as 0
data['ABS_(Anti-lock_Braking_System)']=data['ABS_(Anti-lock_Braking_System)'].replace({'Yes':1,'Not Available':0})
#'Hill_Assist' column
data['Hill_Assist']=data['Hill_Assist'].replace({'Yes':1,'Not Available':0})
#'Central_Locking' column
data['Central_Locking']=data['Central_Locking'].replace({'Yes':1,'Not Available':0})
#'Child_Safety_Locks' column 
data['Child_Safety_Locks']=data['Child_Safety_Locks'].replace({'Yes':1,'Not Available':0})
#'High_Speed_Alert_System' column
data['High_Speed_Alert_System']=data['High_Speed_Alert_System'].replace({'Yes':1,'Not Available':0})
#'Fasten_Seat_Belt_Warning' column replaced
data['Fasten_Seat_Belt_Warning']=data['Fasten_Seat_Belt_Warning'].replace({'Yes':1,'Not Available':0})
#'Door_Ajar_Warning' column replaced
data['Door_Ajar_Warning']=data['Door_Ajar_Warning'].replace({'Yes':1,'Not Available':0})
```

```python
dt=data.iloc[:][['Make','Model','Seating_Capacity','Central_Locking','Child_Safety_Locks','Hill_Assist','High_Speed_Alert_System','ABS_(Anti-lock_Braking_System)','Airbags','Door_Ajar_Warning','Fasten_Seat_Belt_Warning']]
dt
```

```python
#task1: safety feature assessment
for val in dt['Airbags']:
    if val=='Not Available':
        dt['Airbags_avail']=0
    else:
        dt['Airbags_avail']=1

dt
```

```python
dt1=dt.iloc[:][['Make','Model','ABS_(Anti-lock_Braking_System)','Airbags_avail','Hill_Assist']]
dt1.insert(5,'safety',(dt1['ABS_(Anti-lock_Braking_System)']+dt1['Airbags_avail']+dt1['Hill_Assist']))
dt1
```

```python
#Task2: user comfort exploration
dt2=pd.pivot_table(data,index='Make',values='Seating_Capacity',aggfunc='mean')
dt2
```

```python
dt3=dt.iloc[:][['Make','Model','Central_Locking','Child_Safety_Locks']]
dt3.insert(4,'comfort',(dt3['Central_Locking']+dt3['Child_Safety_Locks']))
dt3
```

```python
#task 3
dt4=data.iloc[:][['S. No.','Make','Model','High_Speed_Alert_System','Fasten_Seat_Belt_Warning','Door_Ajar_Warning']]
dt4.insert(6,'Alert',(dt4['High_Speed_Alert_System']+dt4['Fasten_Seat_Belt_Warning']+dt4['Door_Ajar_Warning']))
dt4
```

```python
#task4: dimension analysis
pvt_dim=data.groupby('Make')[['Height_mm','Length_mm','Width_mm']].mean().sort_values(['Height_mm','Length_mm','Width_mm'],ascending=False)
pvt_dim
```

```python
dt1.to_csv('safety.csv')
dt2.to_csv('seat_mean.csv')
dt3.to_csv('comfort.csv')
dt4.to_csv('alert.csv')
pvt_dim.to_csv('dimnsn.csv')
```
